**** For the following columns ['Designation','JobCity','10board','12board','Specialization','CollegeState'], we would be removing the leading spaces and trailing spaces and for uniformity we would be keeping all of them in lower case:

```python
def correction(col):
    df[col]=df[col].str.strip().str.lower()
```

```python
columns = ['Designation','JobCity','10board','12board','Specialization','CollegeState']
```

```python
for col in columns:
    value=df[col].nunique()
    print(f'Number of unique values in {col} is {value}')
```

**** We need to reduce the category as it contains lot of distinct values. We can exclude College State and Specialization from this. 

```python
def reduce(col):
    for value in df[col].unique():
        min_value= df[col].value_counts()[:15].min()
        if df[df[col] == value][col].value_counts()[0] < min_value:
            df.loc[df[col] == value, col] = 'other'
```

```python
for col in columns:
    reduce(col)
```

```python
for col in columns:
    print('Top 10 categories in:', col)
    print('----')
    print(df[col].value_counts())
    print('-'*100)
```

### Feature Engineering:

```python
delta = (df['DOL'] - df['DOJ'])
tenure = np.zeros(len(df))
for i, date in enumerate(delta):
    tenure[i] = round(date.days/365,2)
df['Tenure'] = tenure
```

```python
def cdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1)/len(x)
    return x, y
```

```python
len(df[(df['GraduationYear'] > df['DOJ'].dt.year)].index)
df = df.drop(df[(df['GraduationYear'] > df['DOJ'].dt.year)].index)
```

## 1.1 Tenure:

```python
# Summary Plot

plt.figure(figsize=(5, 4))
df['Tenure'].describe()[1:].plot(alpha = 0.9,
                                 marker = 'X', markersize = 8)
plt.title('Statistics for Tenure')
plt.xlabel('Statistical Measures')
plt.tight_layout()
plt.show()

#Box Plot

plt.figure(figsize=(5, 4))
sns.boxplot(df['Tenure'])
plt.xlabel('Tenure')
plt.tight_layout()
plt.show()

# CDF

plt.figure(figsize=(5, 4))
x_tenure, y_tenure = cdf(df['Tenure'])
x_sample_tenure, y_sample_tenure = cdf(np.random.normal(df['Tenure'].mean(), df['Tenure'].std(), size = len(df['Tenure'])))
plt.plot(x_tenure, y_tenure, linestyle = 'None',
        marker = '.', color = 'blue',
         alpha = 0.7, label = 'Tenure')
plt.plot(x_sample_tenure, y_sample_tenure, linestyle = 'None',
        marker ='.', color = 'green',
        alpha = 0.7, label = 'Normal Distribution')
plt.xlabel('Tenure')
plt.ylabel('CDF')
plt.legend()
plt.tight_layout()
plt.show()


#Histogram
plt.figure(figsize = (10,5))
plt.hist(df['Tenure'],
        ec = 'k',
        bins = np.arange(0, df['Tenure'].max()+0.5, 0.5),
        color = 'blue',
        alpha = 0.5,
        label = f"Skewness : {round(df['Tenure'].skew(),2)}",
        density = True)
plt.xticks(ticks = np.arange(0, df['Tenure'].max()+0.5, 0.5))
plt.xlabel('Experience')
plt.ylabel('Density')
plt.axvline(df['Tenure'].mean(), label = f"Mean: {round(df['Tenure'].mean(),2)}",
            linestyle = '--',
           color = 'yellow', linewidth = 2)
plt.axvline(df['Tenure'].median(), label = f"Median: {round(df['Tenure'].median(),2)}",
            linestyle = ':',
           color = 'k', linewidth = 2)
plt.axvline(df['Tenure'].mode()[0], label = f"Mode: {round(df['Tenure'].mode()[0],2)}"
            , linestyle = '-.',
           color = 'red', linewidth = 2)
sns.kdeplot(df['Tenure'])
plt.legend()
plt.show()
```

**Considering all designations at once in Hypothesis**

```python
import random
n = 40
salary_random = random.sample(df3['Salary'].tolist(),n)
print(salary_random)
```

**T_Score**

```python
def t_score(sample_size, sample_mean, pop_mean, sample_std):
    numerator = sample_mean - pop_mean
    denomenator = sample_std / sample_size**0.5
    return numerator / denomenator
```

**Sample Values**

```python
from scipy.stats import t,norm
import statistics

print('Sample Mean: ', statistics.mean(salary_random))
print('Sample Standard Deviation: ', statistics.stdev(salary_random))
```

```python
sample_size = 40
sample_mean = statistics.mean(salary_random)
pop_mean = 275000
sample_std = statistics.stdev(salary_random)
```

**T_value**

```python
t_value = t_score(sample_size, sample_mean, pop_mean, sample_std)
print(t_value)
```

**t_critical**

```python
confidence_level = 0.95

alpha = 1 - confidence_level

t_critical = t.ppf(1 - alpha/2, df = 99)

print(t_critical)
```

**t_test_visualization**

```python
x_min = -200000
x_max = 800000

mean = pop_mean
std = sample_std

x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)

t_critical_left = pop_mean + (-t_critical * std)
t_critical_right = pop_mean + (t_critical * std)

x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='red')

x2 = np.linspace(t_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color='red')

plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.7))
```

```python
if(t_value < t_critical):
    print("Not enough evidence to reject the Null Hypothesis")
else:
    print("There is sufficent evidence to reject the Null Hypothesis")
```

```python
p_value = 2 * (1.0 - norm.cdf(np.abs(t_value)))

print("p_value = ", p_value)

if(p_value > alpha):
    print("There is not enough evidence to reject the Null Hypothesis")
else:
    print("There is sufficent evidence to reject the Null Hypothesis")
```

```python
alpha = 0.05
from scipy.stats import ttest_1samp

prog_analyst_salaries = df3.loc[df3['Designation'] == 'programmer analyst', 'Salary'].values
software_eng_salaries = df3.loc[df3['Designation'] == 'software engineer', 'Salary'].values
hardware_eng_salaries = df3.loc[df3['Designation'] == 'hardware engineer', 'Salary'].values
assoc_eng_salaries = df3.loc[df3['Designation'] == 'associate engineer', 'Salary'].values

expected_range = (250000, 300000)

for job, salaries in [("programmer analyst", prog_analyst_salaries),
                      ("software engineer", software_eng_salaries),
                      ("hardware engineer", hardware_eng_salaries),
                      ("associate engineer", assoc_eng_salaries)]:

    t_stat, p_val = ttest_1samp(salaries, expected_range[0], alternative='greater')

    print(f"One-sample t-test for {job}:")
    print(f"  t_critical: {t_stat:.2f}")
    print(f"  p_value: {p_val:.5e}")

    if p_val < 0.05:
        print("  Result: There is sufficent evidence to reject the Null Hypothesis\n")
    else:
        print("  Result: There is not enough evidence to reject the Null Hypothesis\n")
```
